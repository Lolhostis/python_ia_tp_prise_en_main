```python
#QUESTION 3
%matplotlib notebook
def moy(imgs):
    img_moy=np.zeros(imgs[0].shape)
    for curr_img in imgs:
        img_moy=img_moy+curr_img
    img_moy=img_moy/len(imgs)
    return img_moy


img_moy = moy(imgs)
plt.title('Image moyenne')
plt.imshow(img_moy, cmap='gray')

plt.show()
```
