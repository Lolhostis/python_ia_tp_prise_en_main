```python
%matplotlib notebook 
import numpy as np
import matplotlib.image as im
import matplotlib.pyplot as plt

I=im.imread('pomme.jpeg')
print(I)
plt.imshow(I, cmap='gray') #On lui prÃ©cise comment interpreter le seul canal dÃ©crit dans l'image
```

```python
%matplotlib notebook 
print("Taille de l'image I :", I.shape)
verif_I=True
for i in range(I.shape[0]):
    for j in range(I.shape[1]):
        if I[i][j]>255:
            verif_I=False
            break
if not verif_I:
    print("L'image n'est pas en niveau de gris")
else:
    print("L'image est en niveau de gris")
```

```python
#QUESTION 5
print("I =", I)
I=I.reshape((I.shape[0])*(I.shape[1]), 1)
print("I en colonne:", I)
```

```python
#QUESTION 6
%matplotlib notebook

I_hist=I.reshape(1, (I.shape[0])*(I.shape[1]))[0].copy()
print("I_hist :", I_hist)

plt.hist(I_hist, density=True,color='m')
```

```python
#QUESTION 7
I_hist = I_hist-np.mean(I_hist)
print("I_hist :", I_hist)
```

```python
#QUESTION 8
%matplotlib notebook
plt.hist(I_hist, density=True,color='m')
```
